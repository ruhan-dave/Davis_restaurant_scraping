```python
url = "https://www.yelp.com/search?find_loc=Davis%2C+CA&sortby=recommended&start=0"
results = requests.get(url) # the variable we create to store our request.get action
```

```python
soup = BeautifulSoup(results.text, "html.parser")
"""
# The variable we assign BeatifulSoup method  to, which specifies a desired format of results using the HTML parser;
this allows Python to read the components of the page
"""
print(soup.prettify())
```

```python
links = []
names = []
ratings = []
votes = []
cuisines = []
pricings = []
hours = []
```

```python
def scrape_links(soup, links):
    link_iterator = soup.find_all("a", class_="css-1m051bw")
    for i in link_iterator:
        link = "yelp.com" + i["href"] # <a hredf=...> is under the class
        links.append(link)
    return links
```

```python
scrape_links(soup, links)
```

```python
def scrape_names(soup, names):
    link_iterator = soup.find_all("a", class_="css-1m051bw")
    for i in link_iterator:
        name = i["name"]
        names.append(name)
    return names
```

```python
scrape_names(soup, names)
```

```python
def scrape_ratings(soup, ratings):
    rating_iterator = soup.find_all("div", class_="i-stars__09f24__foihJ")
    for r in rating_iterator:
        rating = r["aria-label"]
        ratings.append(rating)
    return ratings
```

```python
scrape_ratings(soup, ratings)
```

```python
def scrape_votes(soup, votes):
    vote_iterator = soup.find_all("span", class_="reviewCount__09f24__tnBk4")
    for v in vote_iterator:
        vote = v.get_text()
        votes.append(vote)
    return votes
```

```python
scrape_votes(soup, votes)
```

```python
cuisines = []
pricings = []

def scrape_cuisines(soup, cuisines):
#    cuisine_iterator = soup.find_all("p", class_="css-1gfe39a")
#    for c in cuisine_iterator:
    cuisine_iterator = soup.find_all("span", class_="css-epvm6")
    for c in cuisine_iterator:
        cuisine = ""
        iterator = c.find_all("p", class_="css-dd1rsv")
        for i in iterator:
            to_add = i.get_text()
            cuisine += f"{to_add},"
        cuisines.append(cuisine)
    cuisines
    return cuisines

def scrape_pricings(soup, pricings):      
    pricing_iterator = soup.find_all("span", class_="priceRange__09f24__mmOuH css-chan6m")
    for p in pricing_iterator:
        price = p.get_text()
        pricings.append(price)
    return pricings
```

```python
scrape_cuisines(soup, cuisines)
```

```python
scrape_pricings(soup, pricings)
```

```python
import time
import random


for i in range(0, 231, 10):
    page = requests.get("https://www.yelp.com/search?find_loc=Davis%2C+CA&sortby=recommended&start="+str(i))
    soup = BeautifulSoup(page.text, 'html.parser')
    time.sleep(random.randint(2,10)) # make an random time delay so yelp doesn't yell at me for scraping their website
    l = scrape_links(soup, links)
    n = scrape_names(soup, names)
    r = scrape_ratings(soup, ratings)
    v = scrape_votes(soup, votes)
    c = scrape_cuisines(soup, cuisines)
    p = scrape_pricings(soup, pricings)
    add = pd.DataFrame(list(zip(n,l,r,v,c,p)), columns=["restaurant_name", "link", "rating", "votes", "cuisine_type", "pricing"])
    data = pd.concat([data, add])
```
