```python
def get_restaurant_names_and_links(text):
    # text is a list of all the <span> elements with class_="css-1pxmz4g". 
    # We want to get the link and name within/underneath each <span> element in the text list.
    links = []
    names = []
    for n in text:
        finda = n.find("a")
        link = finda["href"]
        link = "yelp.com" + link
        links.append(link)
        name = finda["name"]
        names.append(name)
    return links, names
```

```python
def select_ratings(soup):
    stars = []
    ratings = soup.select("div.i-stars__09f24___sZu0")
    for r in ratings:
        print(rating.prettify())
        rate = r["aria-label"]
        stars.append(rate)
    return stars
```

```python
select_ratings(soup)
```

```python
def find_ratings(soup):
    stars = []
    ratings = soup.find_all("div", class_="i-stars__09f24___sZu0")
    for r in ratings:
        print(r.prettify())
        rate = r["aria-label"]
        stars.append(rate)
    return stars
```

```python
find_ratings(soup)
```

```python
def get_counts(soup):
    num_ratings = soup.find_all("span", class_="reviewCount__09f24__3GsGY css-e81eai")
    counts = []
    for n in num_ratings:
        count = n.get_text()
        counts.append(int(count))
    return counts
```

```python
def get_pricing(soup):
    prices = []
    pricing = soup.find_all("span", class_="priceRange__09f24__2GspP")
    for p in pricing:
        price = len(str(p.get_text()))
        prices.append(price)
    return prices
```
